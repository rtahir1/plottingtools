```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
handle=open("IPG3-10_A431-UU1_n52951peps_reproducibility.txt",'r')
```

```python
handle.readline()
x=[]
y=[]
for line in handle:
    row=line.strip().split("\t")
    x.append(float(row[0]))
    y.append(float(row[1]))

handle.close()  
```

```python
print min(x),max(x)
print min(y),max(y)
x=np.array(x)
y=np.array(y)
```

```python
# Calculate the point density
xy=np.vstack([x,y])
z=gaussian_kde(xy)(xy)
```

```python
# Sort the points by density, so that the densest points are plotted last
idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]
```

```python
%matplotlib inline
```

```python
### plt.subplots() is a function that returns a tuple containing a figure and axes object(s). 
### Thus when using fig, ax = plt.subplots() you unpack this tuple into the variables fig and ax. 
### Having fig is useful if you want to change figure-level attributes or save the figure as an image file later.

fig, ax = plt.subplots()
ax.set_xlim(1,72)
ax.set_ylim(1,72)
ax.scatter(x, y, c=z,s=10, cmap="OrRd",edgecolor='')
```

```python
def readInput(filename):
    x=[]
    y=[]
    with open(filename,"r") as handle:
        handle.readline()
        for line in handle:
            row=line.strip().split("\t")
            x.append(float(row[0]))
            y.append(float(row[1]))
            
    x=np.array(x)
    y=np.array(y)       
    
    xy=np.vstack([x,y])
    z=gaussian_kde(xy)(xy)
    
    idx = z.argsort()
    x, y, z = x[idx], y[idx], z[idx]
    return x,y,z
        
```

```python
x1,y1,z1=readInput("IPG3-10_A431-UU1_n52951peps_reproducibility.txt")
```

```python
fig, ax = plt.subplots()
ax.set_xlim(1,72)
ax.set_ylim(1,72)
ax.set_xlabel("A431 peptides IEF fraction number")
ax.set_ylabel("Tissues Set1 peptides IEF fraction number")
ax.set_title("A431 IPG3-10 VS Tissues Set1 IPG3-10")
ax.scatter(x1, y1, c=z1,s=10, cmap="OrRd",edgecolor='')
plt.savefig("A431 IPG3-10 VS Tissues Set1 IPG3-10.png")
```

```python
def readInput2(filename):
    x=[]
    y=[]
    with open(filename,"r") as handle:
        handle.readline()
        for line in handle:
            row=line.strip().split("\t")
            x.append(float(row[0]))
            y.append(float(row[1]))
    
    ymin=min(y)
    ymax=max(y)
    
    x_update=[]
    y_update=[]
    for i in range(len(x)):
        if ymin<=x[i]<=ymax:
            x_update.append(x[i])
            y_update.append(y[i])
    
    x=np.array(x_update)
    y=np.array(y_update)       
    
    xy=np.vstack([x,y])
    z=gaussian_kde(xy)(xy)
    
    idx = z.argsort()
    x, y, z = x[idx], y[idx], z[idx]
    return x,y,z
```

```python
x7,y7,z7=readInput2("A431_IPG3-10_vs_IPG37-49_n55737peps_rep.txt")
```

```python
fig, ax = plt.subplots()
ax.set_xlim(min(y7),max(y7))
ax.set_ylim(min(y7),max(y7))
ax.set_xlabel("experimental pI of peptides from IPG3-10")
ax.set_ylabel("experimental pI of peptides from IPG37-49")
ax.scatter(x7, y7, c=z7,s=10, cmap="Oranges", edgecolor='')
plt.savefig("A431 IPG3-10 VS A431 IPG37-49.oranges.png",dpi=300)
```
